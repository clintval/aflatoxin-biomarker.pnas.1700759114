Determining the Frequency of Trinucleotides in the Lambda EG10 genome (6,382 nt)
---

Reference genome (~6.3kb of the ~48kb modified lambda EG10 genome) is available from [transgenic gpt delta ](http://www.nihs.go.jp/dgm/dgm3/gptdeltainfo.html). We need to slice from 32,835-39,217bp (including overlap for edge case trinucleotides) in our reference genome to compute the trinucleotide context proportions in the true source of our reads.

*Note: Our EG10_corrected.fasta has been corrected for a few germline mutations present in our murine model. It differs slightly from the cited sequence.*

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from siglib.utils import timestamp; timestamp()
```

```python
from siglib.utils import fasta_to_dict, seq_to_fasta

EG10_file = '/home/clint/essigmann_analysis/mouse/data/EG10_corrected.fasta'
EG10_file_slice = '/home/clint/essigmann_analysis/mouse/data/EG10_corrected_slice.fasta'

ref = fasta_to_dict(EG10_file)
ref['EG10'].seq = ref['EG10'].seq[32834:39216]
seq_to_fasta(ref['EG10'], EG10_file_slice)
```

```python
import matplotlib.pyplot as plt
from siglib.jellywrap import jellywrap

notation = 'pyrimidine'
outfile = '/home/clint/essigmann_analysis/mouse/data/EG10_slice_counts.jf'

j = jellywrap()
j.count(EG10_file_slice, outfile, canonical=True, notation=notation)

fig, ax = plt.subplots(figsize=(15, 5))
ax = j.plot(ax=ax, title='Trinucleotide Context Counts in EG10 Genome Slice (6382 nt)')
```
