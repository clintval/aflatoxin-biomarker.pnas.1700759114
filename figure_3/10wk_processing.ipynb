```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import siglib
```

```python
from siglib.pileup import Pileup
from siglib import data_p

pileups = []
min_depth = 100; notation = 'pyrimidine'

for treatment in sorted(data_p[10].keys()):
    for id, infile in sorted(data_p[10][treatment].items()):
        pileups.append(Pileup(infile, id=id, description=treatment))
```

```python
for pileup in pileups:
    pileup.parse(min_depth=min_depth)
    pileup.plot_statistics(title='{} - {} Treated'.format(pileup.id, pileup.description))
```
